```python
import gym

from stable_baselines.common.policies import MlpPolicy
from stable_baselines.common import make_vec_env
from stable_baselines import PPO2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# multiprocess environment
env = make_vec_env('CartPole-v1', n_envs=4)

model = PPO2(MlpPolicy, env, verbose=1)
model.learn(total_timesteps=25000)
model.save("ppo2_cartpole")

del model # remove to demonstrate saving and loading

model = PPO2.load("ppo2_cartpole")

# Enjoy trained agent
obs = env.reset()
while True:
    action, _states = model.predict(obs)
    obs, rewards, dones, info = env.step(action)
    env.render()
```

```python
import sys
sys.path.insert(1, './env')
sys.path.insert(1, './agents')
import gym

from stable_baselines.common.policies import MlpPolicy
from stable_baselines.common import make_vec_env
from stable_baselines import PPO2
import aa_gun

report = []

done_max = 5
resets_max = 140

for estimation in range(done_max):
    # multiprocess environment
    env=aa_gun.AA_gun_simple0_env(fast_planes=True,long_projectiles=True,random_speed=True)
    #env = make_vec_env(env, n_envs=4)

    model = PPO2(MlpPolicy, env, verbose=0)
    model.learn(total_timesteps=100)
    

    # Enjoy trained agent
    
    obs = env.reset()
    episode=0
    rew_episodes=[]
    print('ESIMATION',estimation)
    while True:
        action, _states = model.predict(obs)
        obs, rewards, dones, info = env.step(action)
        reward_list.append(rewards)
        #env.render()
        if dones:
            rew_episodes.append(np.sum(reward_list))
            print('EPISODE',episode,'rew',np.sum(reward_list))
            obs = env.reset()
            reward_list=[]
            episode+=1
        if episode>resets_max:
            break
    report.append(np.mean(rew_episodes))
    plt.plot(rew_episodes)
    plt.show()
    
```
